# Creating N-gram Language Models

In this notebook, we see how the CountVectorizer in scikit-learn can be used 
to create n-gram language models.

Let's start by using a sample corpus to inspect the workings of the CountVectorizer: 

```python
from sklearn.feature_extraction.text import CountVectorizer

corpus = ['\n'.join(['the cat sits on the mat', 'the dog sits on the log'])]

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(corpus)

print(vectorizer.get_feature_names_out())
print(X.toarray())
```

To get bigrams (and so on), CountVectorizer allows you to specify start and stop values for n:

```python
vectorizer = CountVectorizer(ngram_range=(2, 2))
X = vectorizer.fit_transform(corpus)

print(vectorizer.get_feature_names_out())
print(X.toarray())
```

```python
vectorizer = CountVectorizer(ngram_range=(3, 3))
X = vectorizer.fit_transform(corpus)

print(vectorizer.get_feature_names_out())
print(X.toarray())
```

### Creating a language model from the CountVectorizer results

1. Fit the CountVectorizer on your corpus for a given value of $n$
2. Sum across documents to get a corpus-level count of n-grams
3. Create dictionaries with counts:
    - ngram_count[(w1, w2, ...)] = c
    - context_count[(w1, ...)] = total times the prefix (w1, ...) appears
4. Convert dictionaries with counts to dictionaries with probabilities (optionally with smoothing)

### Generating sentences using next word prediction

1. Create a function that takes a sentence
2. Extract the last (n-1) words from the sentence
3. Select a random word from the probability dictionary,  
   using the distribution of probabilities in the dictionary.
4. Add the generated word to the sentence and  
   repeat for a specified number of times
