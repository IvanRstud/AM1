# Python Virtual Environments and Using the Terminal

### Why Use the Terminal?

The terminal is:
- A direct interface to your OS
- Reproducible
- Scriptable
- The native way Python tools work

Open in VS Code: **Terminal → New Terminal** (PowerShell on Windows).

#### Navigating the Directory Tree (Windows Terminal Basics)

Before creating environments, you need to move around your file system.

Think of your computer as a tree:
```text
C:\
 ├── Users
 │    └── Username
 │         └── Downloads
 │              └── Subfolder
 ```

In the terminal, you are always inside a folder. The current folder is usually shown in the prompt:
```powershell
PS C:\Users\Username\Downloads> 
```


You enter **commands** on the prompt, which will be executed by the shell (commonly Command prompt or Powershell).  
You can change the current folder using the `cd` command. The accepts both relative and absolute foldernames:
```powershell
PS C:\Users\Username\Downloads> cd Subfolder   # relative path
PS C:\Users\Username\Downloads\Subfolder> # now in the subfolder

PS C:\Users\Username\Downloads\Subfolder> cd C:\Users\Username   # aboslute path
PS C:\Users\Username>
```

You can go up one level in the directory tree by using `..`, designating the **parent directory**. This also works as part of a longer relative path.
```powershell
PS C:\Users\Username\Downloads> cd ..   # navigate to parent folder
PS C:\Users\Username>

PS C:\Users\Username\Downloads\Subfolder> cd ..\AnotherSubfolder   # relative path using ..
PS C:\Users\Username>Downloads\AnotherSubfolder>
```

### The Problem with Global Python

Run the following in the terminal (not in a notebook cell):

```powershell
PS C:\Users\Username\Downloads> python --version
Python 3.13.7
PS C:\Users\Username\Downloads> (get-command python).Path     
C:\Python313\python.exe
```

Think about:
- Which Python are you running by default?
- Where is it installed?
- What happens if you install packages globally? Where do they go?

If you install packages globally:

```powershell
pip install numpy
```

You modify your system Python.

Problems:
- Version conflicts between projects
- Broken tools
- Non-reproducible environments

Solution: Each project gets its own environment.


### What is a Virtual Environment?

A **virtual environment (venv)** is a self-contained Python installation
with its own packages.

Benefits:
- Dependency isolation
- Clean deletion
- Reproducibility
- Version control per project


### Creating a Virtual Environment

In the VS Code terminal:

#### Step 1 – Create project folder
```powershell
mkdir terminal_demo  # create a new folder for the project
cd terminal_demo     # navigate into the new folder
```

#### Step 2 – Create venv
```powershell
python -m venv .venv  # create a new virtual environment named .venv
```

#### Step 3 – Activate (Windows)
```powershell
.venv\Scripts\activate.ps1  # activate the virtual environment
```

You should now see `(.venv)` in the terminal.

Check:
```powershell
(get-command python).Path
```
It should point inside `.venv`.


#### Step 4 – Install a package

```powershell
pip install requests
pip list
```

You can easily test the installation of the package(s) using Python in the terminal:
```powershell
C:\Users\Username\terminal_demo> python
Python 3.13.7 (tags/v3.13.7:bcee1c3, Aug 14 2025, 14:15:11) [MSC v.1944 64 bit (AMD64)] on win32
Type "help", "copyright", "credits" or "license" for more information.
Ctrl click to launch VS Code Native REPL
>>> import requests
>>> requests.__version__
'2.32.5'
```

Deactivate:
```powershell
deactivate
```

Try running Python and importing the requests package again, now outside the environment.
It should fail if `requests` is not globally installed (but it probably is, since you're using Anaconda).

### Why We Use pip (and Do Not Mix Package Managers)

**pip**:
- Python’s official package manager
- Works naturally with `venv`

**conda**:
- Environment manager + package manager
- Installs Python and system libraries

### The danger of mixing
If you use `pip` inside a conda environment:
- Dependency resolution can break
- Binary incompatibilities may occur
- You may get mysterious crashes or DLL errors

Rule for this course:
```powershell
python -m venv .venv
pip install ...
```
Use one package manager only.


### Recommended Workflow

For every new project:

```powershell
mkdir project_name
cd project_name
python -m venv .venv
.venv\Scripts\activate
pip install <packages>
pip freeze > requirements.txt
```

To recreate the environment later:

```powershell
pip install -r requirements.txt
```


### Caveats

Word of warning:
**Do not create a virtual environment in a folder that is stored in the cloud (OneDrive/Google Drive/etc).**  
The environment will contains thousands of files and is a pain to synchronise.
