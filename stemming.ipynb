### Imports

```python
import os
import json
from collections import Counter

import nltk
from nltk.stem import PorterStemmer

nltk.download("wordnet")
nltk.download("omw-1.4")
nltk.download('punkt')
nltk.download("averaged_perceptron_tagger")
nltk.download('averaged_perceptron_tagger_eng')# POS tagger (download BOTH to be safe across NLTK versions)
```

### Settings

```python
DIR_DATA = "data"
DIR_TEXTS = os.path.join(DIR_DATA, "texts")
FILE_METADATA = os.path.join(DIR_DATA, "metadata.csv")
```

### Stemming

```python
# Load the cleaned corpus
with open(os.path.join(DIR_DATA, "cleaned_corpus.txt"), "r", encoding="utf-8") as f:
    corpus = f.read()
    
tokens = corpus.split() # simple whitespace tokenization for POS tagging
```

```python
# Initialize the Porter Stemmer
stemmer = PorterStemmer()

# Stem all tokens in the corpus
stems = [stemmer.stem(token) for token in tokens]
print(f"Stemming produced {len(set(stems))} unique stems from {len(set(tokens))} unique original tokens.")
```

```python
# Print results
print("Sample of original tokens and their corresponding stems (skipping identical ones):")
n_diff_stems = 0
for word, stem in zip(tokens, stems):
    if word != stem:
        n_diff_stems += 1
        print(f"{word} → {stem}")
        if n_diff_stems >= 10:
            break
```

```python
stem_counter = Counter(stems)
print("Unique tokens after stemming:", len(stem_counter))
print("Most common stems:")
# Inspect common stems
stem_counter.most_common(5)
```

```python
# Store Stemming output
with open("stemmed_tokens.json", "w") as f:
    json.dump(stems, f)
```
